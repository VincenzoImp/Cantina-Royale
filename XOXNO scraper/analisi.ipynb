```python
import json
```

```python
with open('results.json', 'r') as f:
    results = json.load(f)
```

```python
results = {x: results[x] for x in results if results[x]['ipfs']['attributes'][1]['value'] in ['Gold', 'Epic', 'Legendary']}
```

```python
l = sorted(results, key=lambda x: results[x]['verko']['gameData'][0]['dynamicData']['level'], reverse=True)
```

```python
q = sorted(results, key=lambda x: results[x]['verko']['gameData'][0]['dynamicData']['character_tokens'], reverse=True)
```

```python
d = {}
for i in l:
    d[i] = {
        'level': results[i]['verko']['gameData'][0]['dynamicData']['level'],
        'character_tokens': results[i]['verko']['gameData'][0]['dynamicData']['character_tokens'],
        'price': results[i]['price']
    }
```

```python
import matplotlib.pyplot as plt

d = {i: d[i] for i in d if d[i]['price'] < 12 and d[i]['level'] > 9}
plt.figure(figsize=(20, 10))
plt.plot([d[i]['level'] for i in d], [d[i]['price'] for i in d], 'o')
# label each point
for i in d:
    plt.annotate(i, (d[i]['level'], d[i]['price']))
plt.title('Price vs Level')
plt.xlabel('Level')
plt.ylabel('Price')
plt.show()
```
