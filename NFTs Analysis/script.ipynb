```python
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt
```

```python
with open('nfts.json', 'r') as f:
    data = json.load(f)
```

```python
bronze = [ape['level'] for ape in data if ape['rarity'] == 'Bronze']
silver = [ape['level'] for ape in data if ape['rarity'] == 'Silver']
gold = [ape['level'] for ape in data if ape['rarity'] == 'Gold']
epic = [ape['level'] for ape in data if ape['rarity'] == 'Epic']
legend = [ape['level'] for ape in data if ape['rarity'] == 'Legendary']
```

```python
maxlv = max(data, key=lambda x: x['level'])['level']
maxlv
```

```python
P_bronze = np.array([bronze.count(lv) for lv in range(1,maxlv+1)])
P_silver = np.array([silver.count(lv) for lv in range(1,maxlv+1)])
P_gold = np.array([gold.count(lv) for lv in range(1,maxlv+1)])
P_epic = np.array([epic.count(lv) for lv in range(1,maxlv+1)])
P_legend = np.array([legend.count(lv) for lv in range(1,maxlv+1)])
```

```python
plt.figure(figsize=(20, 10), dpi=1000)

x = [str(lv) for lv in range(1,maxlv+1)]

plt.bar(x, P_bronze, color='brown')
plt.bar(x, P_silver, bottom=P_bronze, color='silver')
plt.bar(x, P_gold, bottom=P_bronze+P_silver, color='gold')
plt.bar(x, P_epic, bottom=P_bronze+P_silver+P_gold, color='green')
plt.bar(x, P_legend, bottom=P_bronze+P_silver+P_gold+P_epic, color='purple')
plt.xticks(range(len(x)), x, rotation=90)
plt.draw()
```

```python
l = [ape['level'] for ape in data if ape['level']>1]
plt.hist(l, bins=70, log=True)
plt.show()
```
