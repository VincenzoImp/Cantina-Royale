```python
import requests
import time
import tqdm

collection = 'CEA-2d29f9'
size = 100
initialTimestamp = 1663851636
step = 86400
l = []
now = int(time.time())

for ts in tqdm.tqdm(range(initialTimestamp, now, step)):
    start = 0
    while True:
        response = requests.request(
            method='get',
            url='https://api.multiversx.com/collections/{}/transactions?from={}&size={}&before={}&after={}&order=asc'.format(
                collection, start, size, ts+step-1, ts
                )
            )
        if response.status_code == 200:
            js = response.json()
            if js == []:
                break
            else:
                start += size
                l += js
        else:
            time.sleep(1)
```

```python
name = 'GenCantinaElrondApes'
```

```python
trasactions = sorted([x for x in l if x['status'] == 'success'], key=lambda x: x['timestamp'])
d = dict()
for tx in trasactions:
    txHash = tx['txHash']
    for index, item in enumerate(tx['action']['arguments']['transfers']):
        if item['name'] == name:
            identifier = item['identifier']
            fromAddr = tx['sender']
            toAddr = tx['action']['arguments']['receiver']
            try:
                function = tx['function']
            except KeyError:
                function = 'None'
            tup = (txHash, function, index, fromAddr, toAddr)
            if identifier in d.keys():
                d[identifier].append(tup)
            else:
                d[identifier] = [tup]
```
